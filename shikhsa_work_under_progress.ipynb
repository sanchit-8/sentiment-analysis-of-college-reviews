```python
Website="Shiksha"
Placement=[]
Infra=[]
Facultyr=[]
Or=[]
Rating=[]

Combined=[]
Title=[]
i=0

```

```python
def shiksha(str):
    #print("bbye")
    #print(str)
    
    i=0
    headers={
        "User-Agent" : "Mozilla/5.0"
    }
    r=requests.get(str,headers=headers)
    content=r.content
    parser = BeautifulSoup(content)
   
    data(parser)
    df=pd.DataFrame({"Website":Website,"Title":Title,"Review":Combined,"Placement":Placement,"Infrastructure":Infra,"Faculty":Facultyr,"Social":None,"Other":Or,"Label":None,"Rating":Rating})
    return(df)
```

```python
def data(parser):
    i=0
    for strng in parser.findAll('div', attrs = {'class':'group-card pad-off rvw-crd'}):
        Placement_flag=0
        Infra_flag=0
        Faculty_flag=0
        Other_flag=0
        a=""
        
        i=i+1
        
        if i>1:
            for var in strng.findAll('div', attrs = {'new_rating'}):
                rating=var.span.text[1:4]
                Rating.append(rating)
            for var in strng.find_all('div', attrs = {'class':'new_rating'}):
                ant=var.text
                Title.append(ant[36:])
            for strng1 in strng.findAll('p', attrs = {'class':'rvw-titl'}):
                if strng1.strong==None:
                    Or.append(" ")
                    Other_flag=1
                else:
                    s=strng1.strong.text
                    colhead=s[:-2]

                    foar=strng1.get_text()
                    length=foar.find(":")
                    rev=foar[length+2:]
                    a=a+rev

                    if(colhead=="Placements"):
                        Placement.append(rev)
                        Placement_flag=1
                    if(colhead=="Infrastructure"):
                        Infra.append(rev)
                        Infra_flag=1
                    if(colhead=="Faculty"):
                        Facultyr.append(rev)
                        Faculty_flag=1
                    if(colhead=="Other"):
                        Or.append(rev)
                        Other_flag=1
            
            
            if(Placement_flag==0):
                Placement.append(" ")
                #print("no place")
                Placement_flag=1
            if(Infra_flag==0):
                Infra.append(" ")
                #print("no infra")
                Infra_flag=1
            if(Faculty_flag==0):
                Facultyr.append(" ")
                #print("no fac")
                Faculty_flag=1
            if(Other_flag==0):
                Or.append(" ")
                #print("no other")
                Other_flag=1
            if(a==""):
                for strng1 in strng.findAll('div', attrs = {'class':'tabv_1'}):
                    a=strng1.p.text
                    #print(a)
                    #print("i am here")
            
            Combined.append(a)

    for strng in parser.findAll('div', attrs = {'class':'n-pagination'}):
        for var in parser.findAll('li', attrs = {'class':'next'}):
            str1=var.a['href']
            #print(str1)
    
            shiksha(str1)
```

```python
def callfunc(str1):
    global Placement
    global Infra
    global Facultyr
    global Or
    global Rating
    global Combined
    global Title
        
    Placement=[]
    Infra=[]
    Facultyr=[]
    Or=[]
    Rating=[]
    Combined=[]
    Title=[]
    return shiksha(str1)
```
