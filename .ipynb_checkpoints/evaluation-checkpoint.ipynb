```python
def shiksha(df):
    struct_f={"no_of_pos":0,"no_of_neg":0,"no_of_neu":0,"total_score":0,"count":0,"avg_score":0}
    struct_p={"no_of_pos":0,"no_of_neg":0,"no_of_neu":0,"total_score":0,"count":0,"avg_score":0}
    struct_i={"no_of_pos":0,"no_of_neg":0,"no_of_neu":0,"total_score":0,"count":0,"avg_score":0}
    struct_s={"no_of_pos":0,"no_of_neg":0,"no_of_neu":0,"total_score":0,"count":0,"avg_score":0}
    struct_r={"no_of_pos":0,"no_of_neg":0,"no_of_neu":0,"total_score":0,"count":0,"avg_score":0}

    for i in range(0,len(df)):    
        if(df.iloc[i]['Website']=="Shiksha"):
            l=sia.polarity_scores(df.iloc[i]['Faculty'])['compound']
            #print(l)
            if(l>0):
                struct_f['no_of_pos']=struct_f['no_of_pos']+1
            elif(l<0):
                struct_f['no_of_neg']=struct_f['no_of_neg']+1
            elif(l==0):
                struct_f['no_of_neu']=struct_f['no_of_neu']+1
            struct_f['total_score']=struct_f['total_score']+l
            struct_f['count']=struct_f['count']+1
            
            p=sia.polarity_scores(df.iloc[i]['Placement'])['compound']
            #print(l)
            if(p>0):
                struct_p['no_of_pos']=struct_p['no_of_pos']+1
            elif(p<0):
                struct_p['no_of_neg']=struct_p['no_of_neg']+1
            elif(p==0):
                struct_p['no_of_neu']=struct_p['no_of_neu']+1
            struct_p['total_score']=struct_p['total_score']+p
            struct_p['count']=struct_p['count']+1
            
            infra=sia.polarity_scores(df.iloc[i]['Infrastructure'])['compound']
            #print(l)
            if(infra>0):
                struct_i['no_of_pos']=struct_i['no_of_pos']+1
            elif(infra<0):
                struct_i['no_of_neg']=struct_i['no_of_neg']+1
            elif(infra==0):
                struct_i['no_of_neu']=struct_i['no_of_neu']+1
            struct_i['total_score']=struct_i['total_score']+infra
            struct_i['count']=struct_i['count']+1
            
            soc=sia.polarity_scores(df.iloc[i]['Social'])['compound']
            #print(l)
            if(soc>0):
                struct_s['no_of_pos']=struct_s['no_of_pos']+1
            elif(soc<0):
                struct_s['no_of_neg']=struct_s['no_of_neg']+1
            elif(soc==0):
                struct_s['no_of_neu']=struct_s['no_of_neu']+1
            struct_s['total_score']=struct_s['total_score']+soc
            struct_s['count']=struct_s['count']+1
            
            re=sia.polarity_scores(df.iloc[i]['Review'])['compound']
            #print(l)
            if(re>0):
                struct_r['no_of_pos']=struct_r['no_of_pos']+1
            elif(re<0):
                struct_r['no_of_neg']=struct_r['no_of_neg']+1
            elif(re==0):
                struct_r['no_of_neu']=struct_r['no_of_neu']+1
            struct_r['total_score']=struct_r['total_score']+re
            struct_r['count']=struct_r['count']+1
            
            
    if(struct_f['count']==0):
        struct_f['avg_score']=0
    else:    
        struct_f['avg_score']=struct_f['total_score']/struct_f['count']
    
    if(struct_p['count']==0):
        struct_p['avg_score']=0
    else:    
        struct_p['avg_score']=struct_p['total_score']/struct_p['count']
        
    if(struct_i['count']==0):
        struct_i['avg_score']=0
    else:    
        struct_i['avg_score']=struct_i['total_score']/struct_i['count']
        
    if(struct_s['count']==0):
        struct_s['avg_score']=0
    else:    
        struct_s['avg_score']=struct_s['total_score']/struct_s['count']
        
    if(struct_r['count']==0):
        struct_r['avg_score']=0
    else:    
        struct_r['avg_score']=struct_r['total_score']/struct_r['count']
    
    
    score={"combined":struct_r,
           "faculty":struct_f,
           "Placements":struct_p,
           "Infrastructure":struct_i,
           "Social":struct_s}
      
    #print(score)
    return(score)
```

```python
def evaluate(query):
    
    from pymongo import MongoClient 
    import pandas as pd

    connection = MongoClient()
    database = connection.minor
    cursor= database[query].find()
    df =  pd.DataFrame(list(cursor))
    
    shiksha_df=shiksha(df)
    career_df=career(df)
    dunia_df=dunia(df)
    mouth_df=mouth(df)
    
    data={"shiksha":shiksha_df,"career":career_df,"collegedunia":dunia_df,"mouthshut":mouth_df}
    
    
    client = MongoClient()
    db = client['score']
    x=query[:-16]
    collection = db[x]
    collection.insert_one(data)
    
    return(data)

    
    #l=reviews=df['Review'].apply(lambda x:sia.polarity_scores(x)['compound'])
    #print(l)
```
